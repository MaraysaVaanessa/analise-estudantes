Os resultados indicam que, embora a maioria dos estudantes utilize inteligência artificial, isso não elimina as dificuldades associadas à adaptação ao ensino presencial, sugerindo que a IA atua como ferramenta de apoio, mas não como solução completa.

Durante a análise dos dados coletados, foi possível observar que a maioria dos estudantes já utiliza ferramentas de inteligência artificial em seus estudos. No entanto, ao cruzar o uso da IA com o impacto causado pela volta das atividades presenciais, percebe-se que o uso da tecnologia não elimina as dificuldades enfrentadas pelos alunos.

Os resultados indicam que, entre os estudantes que utilizam inteligência artificial, uma grande parte ainda relata dificuldades na adaptação à rotina presencial, sendo que a maioria aponta que a volta às atividades dificultou sua organização, seja de forma leve ou significativa.

Apesar disso, uma pequena parcela dos alunos afirma que a IA contribuiu positivamente para sua organização, e outros indicam que não houve impacto relevante. Isso demonstra que a inteligência artificial pode atuar como uma ferramenta de apoio no processo de adaptação, mas não é suficiente para resolver todos os desafios enfrentados pelos estudantes.

Além disso, ao considerar fatores como tempo de deslocamento até a instituição, percebe-se que questões externas, como distância e rotina, possuem forte influência no impacto percebido pelos alunos. Dessa forma, os resultados sugerem que as dificuldades enfrentadas não estão relacionadas apenas ao estudo em si, mas também a aspectos logísticos e pessoais.
